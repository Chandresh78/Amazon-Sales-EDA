Exploratory Data Analysis

```python
df.columns
```

```python
#SIZE
ax = sns.countplot(x='Size', data=df)  #ax is just a variable
```

```python
ax = sns.countplot(x='Size', data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

GROUP BY

```python
df.groupby(['Size'], as_index=False)['Qty'].sum().sort_values(by='Qty', ascending=False)
```

```python
S_Qty = df.groupby(['Size'], as_index=False)['Qty'].sum().sort_values(by='Qty', ascending=False)
sns.barplot(x='Size', y='Qty', data=S_Qty)
```

From above graph we can say that most of the people buy "M" Size Apparel.

COURIER STATUS

```python
sns.countplot(data=df, x='Courier Status', hue='Status')
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df, x='Courier Status', hue='Status')
plt.show()
```

```python
#Histogram
df['Size'].hist()
```

```python
df['Category'] = df['Category'].astype(str)
column_data = df['Category']
plt.figure(figsize=(10,5))
plt.hist(column_data, bins=30, edgecolor = 'Black')
plt.xticks(rotation=90)
plt.show()
```

From the above graph we can see that the most bought product is "T-shirt"

```python
#Checking B2B Data by Pie Chart
B2B_Check = df['B2B'].value_counts()

#pie chart
plt.pie(B2B_Check, labels=B2B_Check.index, autopct='%1.1f%%')
plt.axis('equal')
plt.show()
```

From above graph, we conclude that 99.3% of buyers are retailers & 0.7% are B2B buyers.

```python
#Scatter Plot
x_data = df['Category']
y_data = df['Size']
plt.scatter(x_data, y_data)
plt.xlabel('Category')
plt.ylabel('Size')
plt.title('Scatter Plot')
plt.show()
```

```python
#Plotting count of Cities by State
plt.figure(figsize=(12,6))
sns.countplot(data=df, x='ship-state')
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distribution of States')
plt.xticks(rotation=90)
plt.show()
```

Top 10 States

```python
top_10_state = df['ship-state'].value_counts().head(10)
plt.figure(figsize=(12,6))
sns.countplot(data=df[df['ship-state'].isin(top_10_state.index)], x = 'ship-state')
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distribution of State')
plt.xticks(rotation=45)
plt.show()
```

Note: From above graph we can see that most of the buyers are from Maharashtra State.

CONCLUSION:
The Data Analysis reveals that the business has a significant customer base in Maharastra State, mainly serves retailers & fulfills order through Amazon. There is also a high demand for T-shirts where M-Size is the most preferred choice among buyers.
